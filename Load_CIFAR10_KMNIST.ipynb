```python
import torch
import torchvision
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np


# Set up data transform to convert images to tensors
transform = transforms.ToTensor()

# Load the CIFAR-10 dataset
cifar10_dataset = datasets.CIFAR10(root='./data', train=True, download=True, transform=transform)
```

```python
def display_images(images, labels):
    fig, axes = plt.subplots(1, len(images), figsize=(20, 5))
    for idx, image in enumerate(images):
        axes[idx].imshow(transforms.ToPILImage()(image))
        axes[idx].set_title(labels[idx])
        axes[idx].axis('off')
    plt.show()
```

```python

# Get random indices
indices = torch.randperm(len(cifar10_dataset))[:10]

# Get random images and labels
random_images = [cifar10_dataset[i][0] for i in indices]
random_labels = [cifar10_dataset.classes[cifar10_dataset[i][1]] for i in indices]  # Getting class names instead of numerical labels

# Display the images
display_images(random_images, random_labels)
```

```python
import torch
import torchvision.datasets as datasets
import torchvision.transforms as transforms
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

# Set up data transform to convert images to tensors
transform = transforms.ToTensor()

# Load the KMNIST dataset
kmnist_dataset = datasets.KMNIST(root='./data', train=True, download=True, transform=transform)
```

```python
def display_images(images, labels):
    fig, axes = plt.subplots(2, 5, figsize=(5, 5))  # 2 rows, 5 columns, 10 inches by 5 inches
    axes = axes.ravel()  # Flatten the array of axes to easily iterate through it
    for idx, image in enumerate(images):
        axes[idx].imshow(transforms.ToPILImage()(image), cmap='gray')
        axes[idx].set_title(labels[idx])
        axes[idx].axis('off')
    plt.tight_layout()  # Adjusts the spacing between subplots
    plt.show()

# Get random indices
indices = torch.randperm(len(kmnist_dataset))[:10]

# Get random images and labels
random_images = [kmnist_dataset[i][0] for i in indices]
random_labels = [str(kmnist_dataset[i][1]) for i in indices]

# Display the images
display_images(random_images, random_labels)
```
